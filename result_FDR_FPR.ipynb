```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
root = "./results/TE chemical process/"
```

```python
paths = [root + "res_12_0", root + "res_12_1", root + "res_12_11", root + "res_12_2"]
```

```python
FDRs = []
FPRs = []
for path in paths:
    FDR = []
    FPR = []
    for i in range(1, 6):
        F = np.loadtxt(path + "/FDR_" + str(i) + ".txt")
        P = np.loadtxt(path + "/FPR_" + str(i) + ".txt")
        FDR.append(F.tolist())
        FPR.append(P.tolist())
    FDRs.append(FDR)
    FPRs.append(FPR)
```

```python
print(np.array(FDRs).shape)
```

```python
FDRs_avg = np.mean(FDRs, axis=1)
FDRs_std = np.std(FDRs, axis=1)
```

```python
print(FDRs_avg)
print(FDRs_std)
np.save
```

```python
FPRs_avg = np.mean(FPRs, axis=1)
FPRs_std = np.std(FPRs, axis=1)
```

```python
print(FPRs_avg)
print(FPRs_std)
```

```python
np.savetxt("./results/" + "FDRs_avg.txt", FDRs_avg)
np.savetxt("./results/" + "FDRs_std.txt", FDRs_std)
np.savetxt("./results/" + "FPRs_avg.txt", FPRs_avg)
np.savetxt("./results/" + "FPRs_std.txt", FPRs_std)
```

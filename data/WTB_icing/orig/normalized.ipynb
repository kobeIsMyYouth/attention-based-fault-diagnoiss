```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./labeled/15_data.csv")
```

```python
data
```

```python
titles = data.columns
for clo in titles:
    if clo not in ["time", "group", "frozen"]: 
        mean = np.mean(data[clo].values)
        std = np.std(data[clo].values)
        data[clo] = (data[clo].values - mean) / std
        print(clo)
```

```python
data
```

```python
data.to_csv("labeled-normaled/15_data.csv")
```

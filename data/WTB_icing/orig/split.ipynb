```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./labeled/21_data.csv")
```

```python
data
```

```python
indexes = [0]
for i in range(data.values.shape[0]-1):
    if data['frozen'][i] == 1 and data['frozen'][i+1] == 0:
        indexes.append(i+1)

indexes.append(data.values.shape[0])
```

```python
print(indexes)
```

```python
for i in range(len(indexes)-1):
    temp = data[indexes[i]:indexes[i+1]]
    temp.to_csv("./test/21_" + str(i) +".csv", index=None, header=None)
```

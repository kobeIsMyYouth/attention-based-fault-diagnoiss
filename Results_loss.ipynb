```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
##### 定义图片上的字体
font = {'family': 'Times New Roman',
         'style': 'normal',
         'weight': 'bold',
        #'color':  'darkred', 
        'size': 12,
        }
matplotlib.rc('font', **font)
```

```python
train_loss_12_0 = np.loadtxt("./results/TE chemical process/res_12_0/trainlosss.txt")
train_loss_12_1 = np.loadtxt("./results/TE chemical process/res_12_1/trainlosss.txt")
train_loss_12_2 = np.loadtxt("./results/TE chemical process/res_12_2/trainlosss.txt")
train_loss_12_11 = np.loadtxt("./results/TE chemical process/res_12_11/trainlosss.txt")

train_acc_12_0 = np.loadtxt("./results/TE chemical process/res_12_0/train_accs.txt")
train_acc_12_1 = np.loadtxt("./results/TE chemical process/res_12_1/train_accs.txt")
train_acc_12_2 = np.loadtxt("./results/TE chemical process/res_12_2/train_accs.txt")
train_acc_12_11 = np.loadtxt("./results/TE chemical process/res_12_11/train_accs.txt") 

test_loss_12_0 = np.loadtxt("./results/TE chemical process/res_12_0/testlosss.txt")
test_loss_12_1 = np.loadtxt("./results/TE chemical process/res_12_1/testlosss.txt")
test_loss_12_2 = np.loadtxt("./results/TE chemical process/res_12_2/testlosss.txt")
test_loss_12_11 = np.loadtxt("./results/TE chemical process/res_12_11/testlosss.txt")

test_acc_12_0 = np.loadtxt("./results/TE chemical process/res_12_0/test_accs.txt")
test_acc_12_1 = np.loadtxt("./results/TE chemical process/res_12_1/test_accs.txt") 
test_acc_12_2 = np.loadtxt("./results/TE chemical process/res_12_2/test_accs.txt") 
test_acc_12_11 = np.loadtxt("./results/TE chemical process/res_12_11/test_accs.txt") 
```

```python
index = [1,2,3,4]
means = []
stds = []
means.append(np.mean(test_acc_12_0[180::]))
means.append(np.mean(test_acc_12_1[180::]))
means.append(np.mean(test_acc_12_11[180::]))
means.append(np.mean(test_acc_12_2[180::]))
stds.append(np.std(test_acc_12_0[180::]))
stds.append(np.std(test_acc_12_1[180::]))
stds.append(np.std(test_acc_12_11[180::]))
stds.append(np.std(test_acc_12_2[180::]))
```

```python
print(means)
print(stds)
```

```python
xlabels = ["M(3$\sigma$,None)","M(3$\sigma$,$L_1$)","M(3$\sigma$,$L_2$)","M(3$\sigma$,[$L_1,L_2$])"]
plt.figure(figsize=(10, 6))
plt.bar(index, means,yerr = stds, error_kw = {'ecolor' : '0.2', 'capsize' :6})
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.ylim(90, 98)
plt.ylabel('Accuracy')
plt.xlabel('Models')
plt.xticks([1,2,3,4], xlabels)

plt.savefig("./results/TE chemical process/zhifangtu1.svg")
plt.show()
```

```python
test_acc_12_2_1 = np.loadtxt("./results/TE chemical process/res_12_2_1/test_accs.txt")
test_acc_12_2_2 = np.loadtxt("./results/TE chemical process/res_12_2_2/test_accs.txt") 
test_acc_12_2 = np.loadtxt("./results/TE chemical process/res_12_2/test_accs.txt") 
test_acc_12_2_4 = np.loadtxt("./results/TE chemical process/res_12_2_4/test_accs.txt") 
test_acc_12_2_5 = np.loadtxt("./results/TE chemical process/res_12_2_5/test_accs.txt")

index = [1,2,3,4,5]
means = []
stds = []
means.append(np.mean(test_acc_12_2_1[180::]))
means.append(np.mean(test_acc_12_2_2[180::]))
means.append(np.mean(test_acc_12_2[180::]))
means.append(np.mean(test_acc_12_2_4[180::]))
means.append(np.mean(test_acc_12_2_5[180::]))
stds.append(np.std(test_acc_12_2_1[180::]))
stds.append(np.std(test_acc_12_2_2[180::]))
stds.append(np.std(test_acc_12_2[180::]))
stds.append(np.std(test_acc_12_2_4[180::]))
stds.append(np.std(test_acc_12_2_5[180::]))
```

```python
xlabels = ["M(1$\sigma$,$[L_1,L_2]$)","M(2$\sigma$,$[L_1,L_2]$)","M(3$\sigma$,$[L_1,L_2]$)","M(4$\sigma$,$[L_1,L_2]$)", "M(5$\sigma$,$[L_1,L_2]$)"]
plt.figure(figsize=(10, 6))
plt.bar(index, means,yerr = stds, error_kw = {'ecolor' : '0.2', 'capsize' :6})
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.ylim(90, 99)
plt.ylabel('Accuracy (%)')
plt.xlabel('Models')
plt.xticks([1,2,3,4,5], xlabels)

plt.savefig("./results/TE chemical process/zhifangtu2.svg")
plt.show()
```
